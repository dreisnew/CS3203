```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("data.csv")

# Data Overview
print("Data Overview:")
print(data.head())

# Data Distribution
plt.figure(figsize=(14, 10))
plt.subplot(3, 3, 1)
sns.histplot(data=data, x="PH", hue="Season", kde=True)
plt.title("Distribution of Plant Height (PH)")

plt.subplot(3, 3, 2)
sns.histplot(data=data, x="IFP", hue="Season", kde=True)
plt.title("Distribution of Insertion of First Pod (IFP)")

plt.subplot(3, 3, 3)
sns.histplot(data=data, x="NLP", hue="Season", kde=True)
plt.title("Distribution of Number of Stems (NLP)")

plt.subplot(3, 3, 4)
sns.histplot(data=data, x="NGP", hue="Season", kde=True)
plt.title("Distribution of Number of Legumes per Plant (NGP)")

plt.subplot(3, 3, 5)
sns.histplot(data=data, x="NGL", hue="Season", kde=True)
plt.title("Distribution of Number of Grains per Plant (NGL)")

plt.subplot(3, 3, 6)
sns.histplot(data=data, x="NS", hue="Season", kde=True)
plt.title("Distribution of Number of Grains per Pod (NS)")

plt.subplot(3, 3, 7)
sns.histplot(data=data, x="MHG", hue="Season", kde=True)
plt.title("Distribution of Thousand Seed Weight (MHG)")

plt.subplot(3, 3, 8)
sns.histplot(data=data, x="GY", hue="Season", kde=True)
plt.title("Distribution of Grain Yield (GY)")

plt.tight_layout()
plt.show()

# Relationship Between Variables
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.scatterplot(data=data, x="PH", y="GY", hue="Season")
plt.title("Plant Height vs. Grain Yield")

plt.subplot(1, 2, 2)
sns.scatterplot(data=data, x="IFP", y="GY", hue="Season")
plt.title("Insertion of First Pod vs. Grain Yield")

plt.tight_layout()
plt.show()

# Seasonal Comparison
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
sns.boxplot(data=data, x="Season", y="PH")
plt.title("Plant Height by Season")

plt.subplot(2, 2, 2)
sns.boxplot(data=data, x="Season", y="IFP")
plt.title("Insertion of First Pod by Season")

plt.subplot(2, 2, 3)
sns.boxplot(data=data, x="Season", y="MHG")
plt.title("Thousand Seed Weight by Season")

plt.subplot(2, 2, 4)
sns.boxplot(data=data, x="Season", y="GY")
plt.title("Grain Yield by Season")

plt.tight_layout()
plt.show()
```
